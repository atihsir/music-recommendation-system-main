# Music Recommendation System For Meditation & Relaxation


## Understanding the Dataset

### Dataset Overview

The dataset contains information related to Spotify tracks used for building a music recommendation system tailored for meditation and relaxation purposes. It includes a comprehensive set of attributes that describe various aspects of each track.

### Dataset Columns

1. **user_id**: Unique identifier for Spotify users.
   
2. **Track Name**: Name of the track.

3. **Artist Name(s)**: Name of the artist(s) associated with the track.

4. **Album Name**: Name of the album to which the track belongs.

5. **Album Release Date**: Release date of the album.

6. **Track Number**: Unique track number within the album.

7. **Track Duration (ms)**: Duration of the track in milliseconds.

8. **Explicit**: Indicator if the track contains explicit content.

9. **Popularity**: Measure indicating how popular a track is based on its play count.

10. **Added By**: User who added the track to a playlist.

11. **Added At**: Timestamp when the track was added to the playlist.

12. **Danceability**: Metric describing how suitable a track is for dancing based on tempo, rhythm stability, beat strength, and overall regularity.

13. **Energy**: Perceptual measure of intensity and activity in the track.

14. **Key**: The key in which the track is performed. Integers map to pitches using standard Pitch Class notation.

15. **Loudness**: Loudness of the track in decibels (dB).

16. **Mode**: Indicates the modality (major or minor) of the track.

17. **Speechiness**: Presence of spoken words in the track, measured as the ratio of speech-like sounds to music-like sounds.

18. **Acousticness**: Confidence measure from 0.0 to 1.0 indicating whether the track is acoustic.

19. **Instrumentalness**: Predicts whether a track contains vocals. Higher values suggest instrumental tracks.

20. **Liveness**: Detects the presence of an audience in the recording. Higher values represent higher probability of a live audience.

21. **Valence**: Measure from 0.0 to 1.0 describing the musical positiveness conveyed by a track.

22. **Tempo**: Beats per minute (BPM) of the track.

23. **Time Signature**: Estimated overall time signature of the track.

### Usage Context

This dataset is pivotal for developing a music recommendation system that caters specifically to meditation and relaxation needs. By analyzing attributes such as acousticness, instrumentalness, and valence, the system aims to recommend tracks that align with calming and soothing musical preferences.

# Finding Similarities of Users by Pivot Table

````python
music_pivot=music.pivot_table(index='user_id',columns='Track_Name', values='Popularity').reset_index(drop=True)
music_pivot
````

````python
# Impute those NaNs with 0 values
music_pivot.fillna(0,inplace=True)
music_pivot
````

````python
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine,correlation
````

````python
# Calculating Cosine Similarity between Users on array data
music_sim=1-pairwise_distances(music_pivot.values,metric='cosine')
music_sim
````

````python
# Store the results in a dataframe format
music_sim1=pd.DataFrame(music_sim)
music_sim1
````

````python
# Set the index and column names to music ids
#similarity of each users individually
music_sim1.index=music['user_id'].unique()
music_sim1.columns=music['user_id'].unique()
music_sim1
````

````python
#every user is similar to themselves so not considering it
# Nullifying diagonal values
np.fill_diagonal(music_sim,0)
music_sim1
````
